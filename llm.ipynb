list

```python
llm=list(range(5,30,5))#最小，最大，长度
print(llm)
print("The list length={}".format(len(llm)))
print("Maximum value={}".format(max(llm)))
print("Minimum value={}".format(min(llm)))
```

```python
llm=list(map(chr,range(4,300)))#4到300经映射分别赋予不同的值
print(llm)
print("[3:6]->{}".format(llm[3:6]))#3到6位次数，不包括第3位
print("[-3:-1]->{}".format(llm[-3:-1]))#倒数1到2，不包括倒数第1位
print("[-3:]->{}".format(llm[-3:]))#倒数3位
print("[:3]->{}".format(llm[:3]))#前3位
print("[:]->{}".format(llm[:]))#全部
```

```python
print(llm)
print("[2:8:2]->{}".format(llm[2:8:2]))#最小位次（不包括），最大位次，间距
print("[::5]->{}".format(llm[::5]))
print("[3:3:-2]->{}".format(llm[3:3:-2]))
print("[10:3:-2]->{}".format(llm[10:3:-2]))
print("[4::-2]->{}".format(llm[4::-2]))
print("[:6:-2]->{}".format(llm[:6:-2]))
```

```python
print(llm)
llm[5]=88 #第6位次赋值88
print("llm[5]=88->{}".format(llm))
wudi=llm+[None]*8#增加8个none
print("llm+[None]*8->{}".format(wudi))
wudi[14]=2021#第15位次赋值2021
print("wudi[14]=2021->{}".format(wudi))
wudi[-7:-2]=list(range(100,104,2))
print("wudi[-7:-2]=list(range(100,104,2))->{}".format(wudi))
wudi[1:1]=[0,0,0,12]
print("wudi[1:1]=[0,0,0,12]->{}".format(wudi))
del wudi[-2:]
print("del wudi[-2:]->{}".format(wudi))
```

```python
llm=list(map(chr,range(95,100)))
print(llm)
llm.append(94)
print("llm.append(94)->{}".format(llm))
llm.append(list(range(40,60,5)))
print("llm.append(list(range(40,60,5)))->{}".format(llm))
wudi=['*',')','*']
llm.extend(wudi)
print("llm.extend(wudi)->{}".format(llm))
print("llm.count('*')={}".format(llm.count('*')))
print("llm.index('a')={}".format(llm.index('a')))
llm.insert(2,[1000,1200,1500])
print("llm.insert(2,[1000,1200,1500])->{}".format(llm))
print("llm.pop(7)_popup->{}".format(llm.pop(7)))
print("llm.pop(7)_retention->{}".format(llm))
llm.remove('*')
print("llm.remove('*')_retention->{}".format(llm))
llm=[3,42,6,95,4,2]
llm.sort()
print("llm.sort()->{}".format(llm))
```
